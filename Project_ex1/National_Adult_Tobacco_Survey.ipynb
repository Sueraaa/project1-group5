```python
# Dependencies
import pandas as pd
from pathlib import Path
```

```python
# Store filepath in a variable
file = Path("Resources /National_Adult_Tobacco_Survey__NATS_.csv")
```

```python
# Read our data file with the Pandas library
tobacco_servey_df = pd.read_csv(file)
tobacco_servey_df
```

```python
tobacco_servey_df.columns
```

```python
# Reorganising the columns using double brackets
organised_df = tobacco_servey_df[["Year","TopicDesc","MeasureDesc","Gender","Age", "Education"]]
organised_df
```

```python
# Using .rename(columns={}) in order to rename columns
renamed_df = organised_df.rename(columns={"TopicDesc":"Topic Description", "MeasureDesc":"User Description"})
renamed_df.head()
```

```python
# using loc on the logic test above only returns rows where the result is True
e_cigratte_data = renamed_df.loc[renamed_df["Topic Description"] == "E-Cigarette Use (Adults)", :]
e_cigratte_data
```
