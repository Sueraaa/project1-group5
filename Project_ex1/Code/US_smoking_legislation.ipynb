```python
#Dependancies 
import pandas as pd 
from pathlib import Path
import plotly.express as px
import folium 
```

```python
#path to data source
file = Path("Resources/CDC_STATE_System_E-Cigarette_Legislation_-_Youth_Access.csv") 
```

```python
#creating data frame
e_cig_leg_df = pd.read_csv(file)  
e_cig_leg_df
```

```python
e_cig_leg_df.columns
```

```python
#creating data frame with useful headings
df = pd.DataFrame(e_cig_leg_df[['GeoLocation','Enacted_Date','Effective_Date','ProvisionDesc']]) 
df.dropna(inplace= True)
df 
```

```python
# Converting date to numeric value
df['Enacted_Date'] = pd.to_datetime(df['Enacted_Date']) 
df['date_numeric'] = df['Enacted_Date'].astype(int) / 10**9  

# Converting geo location into tuple 
df['GeoLocation'] = df['GeoLocation'].str.strip('()').str.split(',').apply(lambda x: tuple(map(float, x)))

df
```

```python
#Creating a map centered around a point
m = folium.Map(location=[37.7749,-100.4194],zoom_start=4) 
```

```python
#initialising indexer variable (x) 
x =1
for i in df["GeoLocation"]: 
    folium.CircleMarker(
        location = i,
        radius=10,
        popup= 'Just checking', 
        x=+ 1
    ).add_to(m)
```
