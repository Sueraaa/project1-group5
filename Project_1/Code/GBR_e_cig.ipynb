```python
#dependancies 
import pandas as pd
import matplotlib as mlp 
import matplotlib.pyplot as plt 
```

```python
#reading e_cig and cig data set 
path = 'Resources/Daily_use_ecig.csv' 
data = pd.read_csv(path)  
path = 'Resources/cig_smoking_UK.csv' 
dataC=pd.read_csv(path)  
df1 = dataC 
df = data
```

```python
#plotting e-cigarette vs cig use  

plt.plot(df['Year'],df['All persons \naged \n16 and over'], color = 'r', label ='All Persons aged 16 and above')
plt.plot(df['Year'], df1['All persons \naged \n16 and over'], label = 'cig users aged 16 and above')
plt.title('E-cigarette users in Great Britian') 
plt.xlabel('Year') 
plt.ylabel('E-cigarette users (%)')  
plt.legend()
plt.show()
```

```python
#plotting e-cig vd cig users among males in the UK 
plt.plot(df['Year'],df['Men \naged \n16 and over'],label='e-cigarette users')  
plt.plot(df['Year'],df1['Men \naged \n16 and over'], label='cigarette users') 
plt.title('Male and Female E-cigarette users in Great Britian') 
plt.xlabel('Year') 
plt.ylabel('Male UK population (%)')  
plt.legend()
plt.show()
```

```python
#plotting e-cig vs cig users among the UK population aged 16-24 
plt.plot(df['Year'],df['All persons \naged \n16 to 24'], label = 'e-cigarette users') 
plt.plot(df['Year'],df1['All persons \naged \n16 to 24'], label='cigarette users') 
plt.title('cigarette vs e-cigarette users among persons aged 16-24') 
plt.xlabel('Year') 
plt.ylabel('UK population aged 16-24 (%)')  
plt.legend()
plt.show()
```

```python
#plotting e-cig vs cig users among UK population aged 25-34 
plt.plot(df['Year'],df['All persons \naged \n25 to 34'], label = 'e-cigarette users') 
plt.plot(df['Year'],df1['All persons \naged \n25 to 34'], label='cigarette users') 
plt.title('cigarette vs e-cigarette users among persons aged 25-34') 
plt.xlabel('Year') 
plt.ylabel('UK population aged 25-34 (%)')  
plt.legend()
plt.show()
```

```python
#plotting e-cig vs cig users among UK population aged 35-49 
plt.plot(df['Year'],df['All persons \naged \n35 to 49'], label = 'e-cigarette users') 
plt.plot(df['Year'],df1['All persons \naged \n35 to 49'], label='cigarette users') 
plt.title('cigarette vs e-cigarette users among persons aged 35-49') 
plt.xlabel('Year') 
plt.ylabel('UK population aged 35-49 (%)')  
plt.legend()
plt.show()
```

```python
#plotting e-cig vs cig use among Females in the United Kingdom 
plt.plot(df['Year'],df['Women \naged \n16 and over'], label ='e-cigarette users') 
plt.plot(df['Year'],df1['Women \naged \n16 and over'], label ='cigarette users') 
plt.title('cigarette vs e-cigarette users among Women in the United Kingdom') 
plt.xlabel('Year') 
plt.ylabel('Female UK population (%)')  
plt.legend()
plt.show()
```

```python
#plotting e-cig use with male and female split
plt.plot(df['Year'],df['Men \naged \n16 and over'],label='Men aged 16 years and above') 
plt.plot(df['Year'],df['Women \naged \n16 and over'], label ='Women aged 16 years and above') 
plt.title('Male and Female E-cigarette users in Great Britian') 
plt.xlabel('Year') 
plt.ylabel('E-cigarette users (%)')  
plt.legend()
plt.show()
```

```python
#plotting e-cig user by age split 
df= data[['Year', 'All persons \naged \n16 to 24','All persons \naged \n25 to 34',
       'All persons \naged \n35 to 49', 'All persons \naged \n50 to 59',
       'All persons \naged \n60 and over']]  
plt.figure(figsize = (15,12))
plt.plot(df['Year'],df['All persons \naged \n16 to 24'], label ='Persons aged 16-24') 
plt.plot(df['Year'],df['All persons \naged \n25 to 34'], label = 'Persons aged 25-34') 
plt.plot(df['Year'], df['All persons \naged \n35 to 49'],label = 'Persons aged 35-49') 
plt.plot(df['Year'], df['All persons \naged \n50 to 59'], label = 'Persons aged 50-59') 
plt.plot(df['Year'], df['All persons \naged \n60 and over'], label = 'Persons aged 60 and above')
plt.title('E-cigarette users based on age') 
plt.xlabel('Year') 
plt.ylabel('e-cigeratte users (%)')  

plt.legend()
plt.show()
```
