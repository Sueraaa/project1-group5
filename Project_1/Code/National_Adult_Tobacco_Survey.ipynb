```python
# Dependencies
import matplotlib.pyplot as plt
import pandas as pd
from pathlib import Path
import numpy as np
```

```python
# Store filepath in a variable
file = Path("../Resources/National_Adult_Tobacco_Survey__NATS__.csv")
```

```python
# Read our data file with the Pandas library
tobacco_servey_df = pd.read_csv(file)
pd.set_option('display.max_columns', None)
tobacco_servey_df
```

```python
tobacco_servey_df.columns
```

```python
tobacco_servey_df['TopicDesc'].unique()
```

```python
# Reorganising the columns using double brackets
organised_df = tobacco_servey_df[["Year", "TopicId","TopicDesc","MeasureDesc", "Data_Value", "Data_Value_Std_Err","Low_Confidence_Limit","High_Confidence_Limit","Gender","Age", "Education", "Race"]]
organised_df
```

```python
# Using .rename(columns={}) in order to rename columns

renamed_df = organised_df.rename(columns={"TopicDesc":"Topic Description", "MeasureDesc":"User Description", "Data_Value":"Data Value in %",
                                          "Data_Value_Std_Err": "Data Value Standard Error","Low_Confidence_Limit" : "Lower Confidence Limit",
                                          "High_Confidence_Limit": "Higher Confidence Limit", "TopicId": "Topic ID"})
                                          
renamed_df.head()
```

```python
# Identify incomplete rows
renamed_df.count()
```

```python
# Drop all rows with missing information
renamed_df = renamed_df.dropna(how='any')
```

```python
# Verify dropped rows
renamed_df.count()
```

```python
renamed_df
```

```python
# using loc on the logic test above only returns rows where the result is True
e_cigratte_data = renamed_df.loc[renamed_df["Topic Description"] == "E-Cigarette Use (Adults)", :]
e_cigratte_data
```

```python
# using loc on the logic test above only returns rows where the result is True
any_tobaco_use_data = renamed_df.loc[renamed_df["Topic Description"] == "Any Tobacco Use (Adults)", :]
any_tobaco_use_data
```

```python
## All data groups in a bar chart
# Count the number of occurances for each topic
total_counts = renamed_df['Topic Description'].value_counts()

# Create a list of colors (you can customize this)
colors = ['blue', 'green', 'red', 'orange', 'purple']

# Create a bar plot
total_counts.plot(kind='bar', color=colors)

# Add labels
plt.title('Adult Tobacco Usage according to the type in United States')
plt.xlabel('Topic Description')
plt.ylabel('Data Value in %')
plt.show()
```

```python
## All data grops in a pie chart
# Create a pie chart
plt.figure(figsize=(8, 8))
colours = ["orange","green","lightblue","yellow","red","purple","pink"]

# Plotting the pie chart
total_counts.plot(kind='pie', autopct='%1.1f%%', shadow=True, startangle=140, colors=colours)

# Add a title
plt.title('Adult Tobacco Usage according to the type in United States')

# Show the pie chart
plt.show()
```

```python
## According to the education level
# Count the number of occurances for each topic
education_counts = renamed_df['Education'].value_counts()

# Create a list of colors 
colors = ['blue', 'green', 'red', 'orange', 'purple']

# Create a bar plot
education_counts.plot(kind='bar', color=colors)

# Add labels
plt.title('Tobacco Usage According to the Education Level in United States')
plt.xlabel('Education')
plt.ylabel('Data Value %')
plt.show()
```

```python
## According to Race
# Count the number of occurances for each topic
education_counts = renamed_df['Race'].value_counts()

# Create a list of colors 
colors = ['blue', 'green', 'red', 'orange', 'purple']

# Create a bar plot
education_counts.plot(kind='bar', color=colors)

# Add labels
plt.title('Tobacco Usage According to the Race in United States')
plt.xlabel('Race')
plt.ylabel('Data Value')
plt.show()
```

```python
## E-cig use according to gender
# Count occurrences of each gender
gender_counts = e_cigratte_data['Gender'].value_counts()

# Create a pie chart
plt.figure(figsize=(8, 8)) 

# Plotting the pie chart
gender_counts.plot(kind='pie', autopct='%1.1f%%', startangle=140, shadow=True, colors=colours)

# Add a title
plt.title('Gender Distribution in E-Cigarette Use in United States')

# Show the pie chart
plt.show()
```

```python
## E-cig use according to the age
# Count occurrences of each gender
gender_counts = e_cigratte_data['Age'].value_counts()

# Create a pie chart
plt.figure(figsize=(8, 8)) 

# Plotting the pie chart
gender_counts.plot(kind='pie', autopct='%1.1f%%', startangle=140,shadow=True, colors=colours)

# Add a title
plt.title('Age Distribution in E-Cigarette Use in United States')

# Show the pie chart
plt.show()
```

```python
## Box plot that shows the distrubution of the data volume for each topic.
# Define selected topics
selected_topics = ['E-Cigarette Use (Adults)', 'Any Tobacco Use (Adults)', 'Cigarette Use (Adults)']

# Filter the DataFrame for the selected topics
selected_df = renamed_df[renamed_df['Topic Description'].isin(selected_topics)]

# Create a box plot with outlier marker face color set to red
plt.figure(figsize=(10, 6)) 
plt.boxplot([selected_df[selected_df['Topic Description'] == topics]['Data Value in %'] for topics in selected_topics], 
            labels=selected_topics, 
            flierprops=dict(markerfacecolor='red'))

# Add labels and title
plt.ylabel('Data Value in %')
plt.xlabel('Topic Description')
plt.title('Distribution of Tobacco Use for Selected Topics')

# Show the box plot
plt.show()
```

```python
## Comparisons 

e_cig_df = renamed_df['Topic Description'] == ['E-Cigarette Use (Adults)']

# Calculate the average 
average_data_value = e_cig_df.groupby('Topic ID')['Data Value in %'].mean()

average_se_value = e_cig_df.groupby('Topic ID')['Data Value Standard Error'].mean()

# Create a scatter plot
plt.figure(figsize=(10, 6)) 

# Plotting the scatter plot
plt.scatter(average_data_value, average_se_value, color='blue', alpha=0.5)

# Add labels and title
plt.title('Comparison of E-Cigarette Use vs Any Tobacco Use (Adults)')
plt.xlabel('E-Cigarette Use (Adults)')
plt.ylabel('Any Tobacco Use (Adults)')

# Show the scatter plot
plt.show()
```
